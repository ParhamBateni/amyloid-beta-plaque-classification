```python
DATA_FOLDER = "data"
DOWNSCALE_SIZE= (224,224)
# Load the labels file
label_names_df = pd.read_csv(f"{DATA_FOLDER}/label_names.csv")
label_to_name = {}
name_to_label = {}
for i,r in label_names_df.iterrows():
    label_to_name[r['Value']] = r['Name']
    name_to_label[r['Name']] = r['Value']
print(label_to_name)
print(name_to_label)
```

```python
 # create the data label folders
os.makedirs(f'{DATA_FOLDER}/labeled_images',exist_ok=True)
os.makedirs(f'{DATA_FOLDER}/unlabeled_images', exist_ok=True)
for name in name_to_label:
   os.makedirs(f'{DATA_FOLDER}/labeled_images/{name}', exist_ok=True)
```

```python
# Display 3 examples of each category of plaques

# Get all label names
label_names = list(name_to_label.keys())

# Create a subplot grid: rows = number of labels, cols = 3 examples
fig, axes = plt.subplots(len(label_names), 3, figsize=(15, 5*len(label_names)))
fig.suptitle('Examples of Each Plaque Category', fontsize=16)

for i, label_name in enumerate(label_names):
    label_folder = f'{DATA_FOLDER}/labeled_images/{label_name}'
    
    # Get list of image files in this category
    if os.path.exists(label_folder):
        image_files = [f for f in os.listdir(label_folder) if f.endswith('.png')]
        
        # Take first 3 images (or less if fewer available)
        for j in range(min(3, len(image_files))):
            if j < len(image_files):
                image_path = os.path.join(label_folder, image_files[j])
                img = plt.imread(image_path)
                
                # Handle single subplot case
                if len(label_names) == 1:
                    ax = axes[j]
                else:
                    ax = axes[i, j]
                
                ax.imshow(img)
                ax.set_title(f'{label_name} - Example {j+1}')
                ax.axis('off')
            else:
                # Handle case where we have fewer than 3 images
                if len(label_names) == 1:
                    ax = axes[j]
                else:
                    ax = axes[i, j]
                ax.text(0.5, 0.5, 'No image', ha='center', va='center', transform=ax.transAxes)
                ax.axis('off')

plt.tight_layout()
plt.show()

# Print summary of images per category
print("\nSummary of images per category:")
for label_name in label_names:
    label_folder = f'{DATA_FOLDER}/labeled_images/{label_name}'
    if os.path.exists(label_folder):
        image_count = len([f for f in os.listdir(label_folder) if f.endswith('.png')])
        print(f"{label_name}: {image_count} images")
    else:
        print(f"{label_name}: 0 images")
```
